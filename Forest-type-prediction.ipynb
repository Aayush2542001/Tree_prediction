```python
import os
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation,Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
```

```python
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train = pd.read_csv('/kaggle/input/tabular-playground-series-dec-2021/train.csv')
test = pd.read_csv('/kaggle/input/tabular-playground-series-dec-2021/test.csv')
```

```python
pd.options.display.max_columns = train.shape[1]
train.describe(include = 'all').apply(lambda s: s.apply('{0:.5f}'.format))
```

# Dropping unncecary columns

```python
train.drop('Soil_Type7', axis=1, inplace=True, errors='ignore')
train.drop('Soil_Type15', axis=1, inplace=True, errors='ignore')
train.drop('Id', axis=1, inplace=True, errors='ignore')
test.drop('Soil_Type7',axis=1, inplace=True, errors='ignore')
test.drop('Soil_Type15', axis=1, inplace=True, errors='ignore')
test_Id = test.Id
test.drop('Id', axis=1, inplace=True, errors='ignore')
```

# Decrementing target column by 1

```python
train.Cover_Type = train.Cover_Type.apply(lambda x: x - 1)
```

# Diving target variable

```python
Y = train.Cover_Type
X = train.drop('Cover_Type', axis=1)
```

# Converting to numpy array

```python
X = X.to_numpy()
Y = Y.to_numpy()
```

# Train_test_split data

```python
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=10)
```

# Model creation

```python
model = Sequential(
    [
        Dense(units=52, input_shape=(52,), activation='relu'),
        Dense(units=7, activation='softmax')
    ]
)
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0005), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=X_train, y=Y_train, validation_data=(X_val,Y_val), batch_size=1000, epochs=40, shuffle=False, verbose=2)
```

# Generating output file

```python
test_prediction = model.predict(test, batch_size=1000)
test_prediction = np.argmax(test_prediction, axis=-1)
test_prediction
```

# Incrementing predicated column by 1

```python
# train.Cover_Type.apply(lambda x: x - 1)
test_prediction = np.vectorize(lambda x: x + 1)(test_prediction)
test_prediction
```

```python
output = pd.DataFrame({'Id': test_Id, 'Cover_Type': test_prediction})
output.to_csv('submission.csv', index=False)
```
